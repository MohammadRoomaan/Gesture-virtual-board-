```python
import cv2
import mediapipe as mp

# Initialize MediaPipe Hands and OpenCV
mp_hands = mp.solutions.hands
hands = mp_hands.Hands()
mp_drawing = mp.solutions.drawing_utils
cap = cv2.VideoCapture(0)

# Canvas to draw on
canvas = None

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Flip the frame horizontally for a mirror-like view
    frame = cv2.flip(frame, 1)
    
    # Initialize the canvas
    if canvas is None:
        canvas = frame.copy()

    # Convert the image to RGB as MediaPipe expects it in RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result = hands.process(rgb_frame)

    # If hand landmarks are detected
    if result.multi_hand_landmarks:
        for hand_landmarks in result.multi_hand_landmarks:
            # Draw hand landmarks on the frame (optional for debugging)
            mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
            
            # Extract coordinates of the index finger tip (landmark #8)
            index_tip_x = int(hand_landmarks.landmark[8].x * frame.shape[1])
            index_tip_y = int(hand_landmarks.landmark[8].y * frame.shape[0])
            
            # Mark the index finger tip with a circle on the frame
            cv2.circle(frame, (index_tip_x, index_tip_y), 10, (0, 255, 0), -1)

            # Use the index finger tip to "draw" on the canvas
            cv2.circle(canvas, (index_tip_x, index_tip_y), 10, (255, 0, 0), -1)

    # Merge the frame with the canvas (show the drawing over the real-time camera feed)
    combined = cv2.addWeighted(frame, 0.5, canvas, 0.5, 0)

    # Show the result
    cv2.imshow('Hand Gesture Drawing', combined)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp

# Initialize MediaPipe Hands and OpenCV
mp_hands = mp.solutions.hands
hands = mp_hands.Hands()
mp_drawing = mp.solutions.drawing_utils
cap = cv2.VideoCapture(0)

# Canvas to draw on
canvas = None

def is_hand_closed(landmarks):
    """Returns True if hand is closed (fist), otherwise False."""
    # Check if the tips of the fingers are near the base joints
    # Thumb (landmark 4), Index (landmark 8), Middle (landmark 12), Ring (landmark 16), Pinky (landmark 20)
    tips = [8, 12, 16, 20]
    bases = [5, 9, 13, 17]
    
    closed = True
    for tip, base in zip(tips, bases):
        # Check if the tip of the finger is below (y-coordinate greater) than the base joint
        if landmarks.landmark[tip].y < landmarks.landmark[base].y:
            closed = False  # Hand is open if any finger tip is above its base joint
            break
    
    return closed

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Flip the frame horizontally for a mirror-like view
    frame = cv2.flip(frame, 1)
    
    # Initialize the canvas
    if canvas is None:
        canvas = frame.copy()

    # Convert the image to RGB as MediaPipe expects it in RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result = hands.process(rgb_frame)

    # If hand landmarks are detected
    if result.multi_hand_landmarks:
        for hand_landmarks in result.multi_hand_landmarks:
            # Draw hand landmarks on the frame (optional for debugging)
            mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
            
            # Extract coordinates of the index finger tip (landmark #8)
            index_tip_x = int(hand_landmarks.landmark[8].x * frame.shape[1])
            index_tip_y = int(hand_landmarks.landmark[8].y * frame.shape[0])

            # Check if the hand is closed
            if not is_hand_closed(hand_landmarks):
                # If the hand is open, draw on the canvas
                cv2.circle(canvas, (index_tip_x, index_tip_y), 5, (0, 0, 0), -1)
            else:
                # If the hand is closed, stop drawing
                cv2.putText(frame, "Hand closed, not drawing", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    # Merge the frame with the canvas (show the drawing over the real-time camera feed)
    combined = cv2.addWeighted(frame, 0.5, canvas, 0.5, 0)

    # Show the result
    cv2.imshow('Hand Gesture Drawing', combined)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp

# Initialize MediaPipe Hands and OpenCV
mp_hands = mp.solutions.hands
hands = mp_hands.Hands()
mp_drawing = mp.solutions.drawing_utils
cap = cv2.VideoCapture(0)

# Canvas to draw on
canvas = None

def is_hand_open(landmarks):
    """Returns True if the hand is fully open (straight fingers), otherwise False."""
    # Check if the tips of the fingers are higher than the middle joints
    tips = [8, 12, 16, 20]
    middles = [6, 10, 14, 18]
    open_hand = True

    for tip, middle in zip(tips, middles):
        if landmarks.landmark[tip].y > landmarks.landmark[middle].y:  # Tip is below middle joint
            open_hand = False  # Hand is not fully open
            break
    
    return open_hand

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Flip the frame horizontally for a mirror-like view
    frame = cv2.flip(frame, 1)
    
    # Initialize the canvas
    if canvas is None:
        canvas = frame.copy()

    # Convert the image to RGB as MediaPipe expects it in RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result = hands.process(rgb_frame)

    # If hand landmarks are detected
    if result.multi_hand_landmarks:
        for hand_landmarks in result.multi_hand_landmarks:
            # Draw hand landmarks on the frame (optional for debugging)
            mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
            
            # Extract coordinates of the index finger tip (landmark #8)
            index_tip_x = int(hand_landmarks.landmark[8].x * frame.shape[1])
            index_tip_y = int(hand_landmarks.landmark[8].y * frame.shape[0])

            # Check if the hand is fully open (straight fingers)
            if is_hand_open(hand_landmarks):
                # If the hand is fully open, enter erase mode (erase with the index finger)
                cv2.circle(canvas, (index_tip_x, index_tip_y), 50, (0, 0, 0), -1)
                cv2.putText(frame, "Erase Mode", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
            else:
                # If the hand is not fully open, continue drawing
                cv2.circle(canvas, (index_tip_x, index_tip_y), 10, (255, 0, 0), -1)

    # Merge the frame with the canvas (show the drawing over the real-time camera feed)
    combined = cv2.addWeighted(frame, 0.5, canvas, 0.5, 0)

    # Show the result
    cv2.imshow('Hand Gesture Drawing and Erasing', combined)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
